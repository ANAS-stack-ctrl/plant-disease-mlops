```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
from collections import Counter
```

```python
DATA_ROOT = r"C:\Users\Asus\plant-disease-mlops\data\raw\PlantVillageDataset\train_val_test"

TRAIN_DIR = os.path.join(DATA_ROOT, "train")
VAL_DIR   = os.path.join(DATA_ROOT, "val")
TEST_DIR  = os.path.join(DATA_ROOT, "test")
```

```python
classes = sorted(os.listdir(TRAIN_DIR))
num_classes = len(classes)

print("Nombre de classes :", num_classes)
print("Classes :")
for c in classes:
    print("-", c)
```

```python
def count_images_per_class(split_dir):
    counts = {}
    for cls in os.listdir(split_dir):
        cls_path = os.path.join(split_dir, cls)
        if os.path.isdir(cls_path):
            counts[cls] = len(os.listdir(cls_path))
    return counts

train_counts = count_images_per_class(TRAIN_DIR)
val_counts   = count_images_per_class(VAL_DIR)
test_counts  = count_images_per_class(TEST_DIR)

print("Train samples:", sum(train_counts.values()))
print("Val samples:", sum(val_counts.values()))
print("Test samples:", sum(test_counts.values()))
```

```python
plt.figure(figsize=(12, 4))
plt.bar(train_counts.keys(), train_counts.values())
plt.xticks(rotation=90)
plt.title("Distribution des images par classe (Train)")
plt.ylabel("Nombre d'images")
plt.tight_layout()
plt.show()
```

```python
def show_random_images(split_dir, n=6):
    all_images = []
    for cls in os.listdir(split_dir):
        cls_path = os.path.join(split_dir, cls)
        for img in os.listdir(cls_path):
            all_images.append((cls, os.path.join(cls_path, img)))

    samples = random.sample(all_images, n)

    plt.figure(figsize=(12, 6))
    for i, (cls, img_path) in enumerate(samples):
        img = Image.open(img_path)
        plt.subplot(2, 3, i+1)
        plt.imshow(img)
        plt.title(cls)
        plt.axis("off")

    plt.tight_layout()
    plt.show()

show_random_images(TRAIN_DIR)
```
